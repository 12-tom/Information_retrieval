### **a)Boolean retrieval**

```python
import os
import nltk
import nltk.corpus
import pandas as pd
import numpy as np
from nltk.tokenize import word_tokenize
```

```python
doc1=r"/content/drive/MyDrive/sample1.txt"
doc2=r"/content/drive/MyDrive/sample2.txt"
doc3=r"/content/drive/MyDrive/sample3.txt"
doc_list=[doc1, doc2, doc3]
import string
token_list=[]
for i in doc_list:
  df=open(i, "r")
  r=df.read()
  punc_r = r.translate(str.maketrans('', '', string.punctuation))
  token_list.append(punc_r.split())
print(token_list)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
print(stop_words)
```

```python
filtered_tokens=[]
unique_tokens=[]
for i in range(0,len(token_list)):
  filtered_tokens.append(i*[])
  for j in token_list[i]:
      if j.lower() not in stop_words:
        filtered_tokens[i].append(j.lower())
  for k in filtered_tokens[i]:
    if k not in unique_tokens:
      unique_tokens.append(k)
print(filtered_tokens)
print(unique_tokens)
```

```python
def vector(token): 
  token_vec=[]
  for i in filtered_tokens:
    if token in i:
      token_vec.append(1)
    else:
      token_vec.append(0)
  return token_vec
```

```python
vector("utility")
```

```python
vector_dict={}
for i in unique_tokens:
    vector_dict[i] = vector(i)

print(vector_dict)
```

```python
query=input("Enter the query:")
query=word_tokenize(query)
operator_words=[]
query_words=[]

for word in query:
    if word.lower() != "and" and word.lower() != "or" and word.lower() != "not":
      query_words.append(word.lower())
    else:
      operator_words.append(word.lower())

if operator_words[0]=="and":
  query_op=np.array(vector_dict[query_words[0]]) & np.array(vector_dict[query_words[1]])
elif operator_words[0]=="or":
  query_op=np.array(vector_dict[query_words[0]]) | np.array(vector_dict[query_words[1]])
elif operator_words[0]=="not":
  query_op=1-np.array(vector_dict[query_words[1]])
  query_op=np.array(vector_dict[query_words[0]]) & query_op
operator_words.pop(0)
query_words.pop(0)
query_words.pop(0)

for word in operator_words:
  if word == "and":
    query_op = query_op & np.array(vector_dict[query_words[0]])
    operator_words.pop(0)
    query_words.pop(0)
  elif word == "or":
    query_op = query_op | np.array(vector_dict[query_words[0]])
    operator_words.pop(0)
    query_words.pop(0)
  elif word == "not":
    query_opt=1-np.array(vector_dict[query_words[0]])
    query_op =query_op & query_opt
    operator_words.pop(0)
    query_words.pop(0)

print(query_op)
doc_dict={'doc1':doc_list[0],'doc2':doc_list[1],'doc3':doc_list[2]}
for i in list(zip(doc_dict,query_op)):
  if i[1]==1:
      print(i[0])
```

## **b)Inverted** **Index**

```python
def freq_token(word):
  cnt=0
  for i in range(0,len(filtered_tokens)):
    for j in filtered_tokens[i]:
      if word==j:
        cnt+=1
  return cnt
```

```python
for i in unique_tokens:
  fq_ind=freq_token(i)
  print(i,':',fq_ind)
```

```python
def ind_token(word):
  inverted_ind=[]
  for i in range(0,len(filtered_tokens)):
    if word in filtered_tokens[i]:
      inverted_ind.append(i)
  return inverted_ind
```

```python
for i in unique_tokens:
  ind=ind_token(i)
  print(i,':',ind)
```

```python
ind_dict={}
for i in unique_tokens:
    ind_dict[i] = {"freq":freq_token(i), "docs":ind_token(i)}

print(ind_dict)
```

```python
def query(qry):
  for i in qry.split():
    if i=='and':
      qry= qry.replace(i,'&')
    elif i=='or':
      qry=qry.replace(i,'|')
    elif i=='not':
      qry=qry.replace(i,"^")
  for i in qry.split():
    if i.isalpha():
      qry=qry.replace(i,repr(set(ind_dict[i]['docs'])))
  return qry
```

```python
qry=input("Enter the query :")
d=eval(query(qry))
if d==set():
  print("No documents")
else:
  for i in d:
    print("doc",i+1,end="")
```
